```python
def psnr(I1,I2):
    """
    Computes the psnr between two images I1 and I2
    """
#     d=np.amax(I1)-np.amin(I1);
    I2 = I2-np.mean(I2);
    d = 2
    diff=np.square(I2-I1);
    MSE=diff.sum()/I1.size;
    Psnr=10*np.log(d**2/MSE)/np.log(10);
    return Psnr
```
