# Import

```python
import torch
import numpy as np
import spyrit.misc.walsh_hadamard as wh

from matplotlib import pyplot as plt

from spyrit.learning.model_Had_DCAN import *
from spyrit.misc.disp import torch2numpy, imagesc, plot
# from spyrit.misc.statistics import Cov2Var
# from spyrit.learning.model_Had_DCAN import Cov2Var
from spyrit.learning.nets import *
from spyrit.misc.metrics import psnr_

def float2_bit_gen(I):
    I_8_bit = 255*(I-np.amin(I))/(np.amax(I)-np.amin(I))
    I_8_bit = I_8_bit.astype('uint8')
    return I_8_bit
```

# Star Sector x12

```python
a_1 = np.load("./results_OE/star_sector_x12_pinv_4096.npy")
a_1_2 = np.load("./results_OE/star_sector_x12_em_net_2048.npy")
a_1_4 = np.load("./results_OE/star_sector_x12_em_net_1024.npy")
a_1_8 = np.load("./results_OE/star_sector_x12_em_net_512.npy")
```

```python
mean_imgs = np.amax(a_1, axis = (1,2))
plot(range(mean_imgs.shape[0]),mean_imgs)
```

```python
Img_1_1 = a_1[500,:,:]
Img_1_2 = a_1_2[500,:,:]
Img_1_4 = a_1_4[500,:,:]
Img_1_8 = a_1_8[500,:,:]
```

```python
imagesc(Img_1_1)
imagesc(Img_1_2)
imagesc(Img_1_4)
imagesc(Img_1_8)
```

```python
Img_1_1 = float2_bit_gen(Img_1_1)
Img_1_2 = float2_bit_gen(Img_1_2)
Img_1_4 = float2_bit_gen(Img_1_4)
Img_1_8 = float2_bit_gen(Img_1_8)
cv2.imwrite('./img/siemens-x12-4096.png', Img_1_1)
cv2.imwrite('./img/siemens-x12-2048.png', Img_1_2)
cv2.imwrite('./img/siemens-x12-1024.png', Img_1_4)
cv2.imwrite('./img/siemens-x12-512.png', Img_1_8)
```
